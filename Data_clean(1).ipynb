But we have  55095 null values in the data, we can drop 34,653 of them but i want to replace them with median.

```python
#changing null values of price to median price based on cab type
ride_dict =  ride.groupby('cab_type')['price'].median().to_dict()
print(ride_dict)
ride['price'] = ride['price'].fillna(ride['cab_type'].map(ride_dict))
```

Lets check for the outliers.

```python

numeric_cols = ride.select_dtypes(include=['int','float'])
num_rows = 9
num_cols = 5
num_plots = num_rows * num_cols
total_plots = len(numeric_cols.columns)
if total_plots > num_plots:
    num_rows = int(np.ceil(total_plots / num_cols))
# Plotting individual box plots for each numeric column
fig, axes = plt.subplots(num_rows, num_cols, figsize=(20, 30))
# Flatten the axes array for easier iteration
axes = axes.flatten()
# Iterate over each numeric column and plot
for i, (col, ax) in enumerate(zip(numeric_cols.columns, axes)):
    if i < total_plots:
        numeric_cols[col].plot(kind='box', ax=ax)
        ax.set_title('Box plot for {}'.format(col))
        ax.set_ylabel('Value')
    else:
        fig.delaxes(ax)  # Remove empty subplots if there are fewer columns than expected
plt.tight_layout()
plt.show()



```

We can see there are bunch of outliers in the predictor variables.

Lets check source and destination count

```python
result  = ride.groupby(['source','destination']).size().reset_index(name='count')
result = result.sort_values(by='count',ascending=False)
result
```

```python
#outliers detection for price
sf = ride['price'].quantile(0.75)
tf = ride['price'].quantile(0.25)
iqr = sf-tf
upper = sf+(1.5*iqr)
lower = sf -(1.5*iqr)
outlier_price = ride[(ride['price']<lower)|(ride['price']>upper)][['source','destination','cab_type']]
outlier_price
count = outlier_price.groupby(by=['source','destination']).value_counts()
index = count.nlargest(10)
index
```

This are the source and destination where cab fare is high ,especially Financial District.
